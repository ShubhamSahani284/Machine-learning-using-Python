```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
dia=pd.read_csv("diabetes.csv")
dia.head()
```

```python
dia.shape
```

```python
dia.info()
```

```python
dia.isnull().sum()
```

```python
sns.heatmap(dia.corr(),annot=True,fmt="0.1f")
```

```python
sns.scatterplot(data=dia, x='Pregnancies', y='Glucose')
```

```python
dia.select_dtypes("object").head()
```

```python
X=dia.iloc[:, :-1]
Y=dia.iloc[:, -1]
```

```python
Y.head()
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split

x1,x2,y1,y2=train_test_split(X,Y,test_size=0.20,random_state=42)
y1.head()
```

```python
y2.head()
```

```python
x1.shape,x2.shape,y1.shape,y2.shape
```

```python
from sklearn.linear_model import LinearRegression

model=LinearRegression()
model.fit(x1,y1)
```

```python
model.score(x2,y2)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
yp=model.predict(x2)
yp
```
